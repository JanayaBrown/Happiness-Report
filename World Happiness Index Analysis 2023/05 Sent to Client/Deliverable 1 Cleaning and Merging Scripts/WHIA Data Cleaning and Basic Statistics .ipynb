# 1. Importing Libraries and Data

```python
# importing libraries
import pandas as pd
import numpy as np
import os
```

```python
# creating main path
path = r'C:\Users\muffi\OneDrive\CareerFoundry\World Happiness Index Analysis 2023'
```

```python
# importing wrangled 2015 - 2019 World Happiness Index data
df_all_wrang = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'all_years_wrangled.csv'), index_col = [0])
df_all_wrang.shape
```

# 2. Missing Data

This section checks for and addresses missing data

```python
# checking for missing values in each col
df_all_wrang.isnull().sum()
```

Missing one value in trust. Since it is less than 5% of the data, I will drop the entry.

```python
# dropping row with missing data
df_all_no_null = df_all_wrang[df_all_wrang['trust_in_government'].isnull() == False]
```

```python
# checking for missing vals
df_all_no_null.isnull().sum()
```

# 3. Duplicate Data

This section checks for and addresses duplicate data

```python
# checking for duplicates
df_all_no_null[df_all_no_null.duplicated()]
```

There are no duplicates

```python
# making post-dup check df
df_all_no_dup = df_all_no_null
```

# Mixed Data

```python
# checking for missing data
for col in df_all_no_dup.columns.tolist():
  weird = (df_all_no_dup[[col]].applymap(type) != df_all_no_dup[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_all_no_dup[weird]) > 0:
    print (col)
```

There are no cols with a mix of data types

```python
# making fully cleaned df
df_all_clean = df_all_no_dup
```

# 4. Exporting Data

```python
# exporting data
df_all_clean.to_csv(os.path.join(path, '02 Data', 'Prepared Data', 'all_years_cleaned.csv'))
```
